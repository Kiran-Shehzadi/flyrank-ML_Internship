## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My Lane as an ML Task

My selected lane is **Content Refresh Prioritization**.

The business goal is to help the SEO team identify which webpages should be refreshed first in order to improve search performance.

I frame this as a **scoring (regression)** problem. Instead of making a simple Yes/No decision, the model assigns each webpage a **refresh priority score**. Pages with higher scores are expected to benefit more from a content update.

The SEO team can then rank pages using these scores and prioritize the highest-value pages for refresh.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The starter dataset does not contain the actual improvement in traffic after refreshing a page, so a direct target is unavailable.

Instead, I would create a **proxy target** called **Refresh Priority Score**.

This score estimates how urgently a page should be refreshed based on signals such as:

- Click-through Rate (CTR)
- Average Position
- Content Age
- Trend Direction
- Impressions

# Higher scores indicate higher refresh priority.

## Success Metric

The success of this ML system would ultimately be measured by business outcomes after refreshing the recommended pages.

Important success metrics include:

- Increase in organic traffic

## Unit of Analysis

The unit of analysis is **one webpage**.

Each row in the dataset represents a single webpage, while each column contains information describing that page, such as:

- Click-through Rate (CTR)
- Impressions
- Average Position
- Content Age
- Trend Direction

These features help determine whether a webpage should be refreshed.

```python
df["refresh_priority_score"] = (
    (1 - df["ctr"]) * 0.4 +
    (df["content_age_days"] / df["content_age_days"].max()) * 0.3 +
    (df["avg_position"] / df["avg_position"].max()) * 0.3
)

df[[
    "ctr",
    "content_age_days",
    "avg_position",
    "refresh_priority_score"
]].head()
```

The **refresh_priority_score** shown above is an example proxy target.

It combines several important signals to estimate which webpages should be refreshed first.

This score is only a demonstration for task framing and is **not** a trained machine learning prediction.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule, such as "refresh every page older than 180 days," ignores many other important factors.

For example:

- An old page may still perform very well.
- A newer page may already be losing traffic.
- Two pages with the same CTR may have very different search positions or impression counts.

Machine learning can evaluate multiple signals together, including CTR, average position, content age, impressions, and trend direction.

By learning patterns across these features, an ML model can produce better refresh recommendations than a single threshold-based rule.
