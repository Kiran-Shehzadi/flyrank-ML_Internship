## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Problem Framing

### Decision

Which pages should an SEO team prioritize for a content refresh?

### Who acts on the output?

The primary user is an SEO specialist or content editor.

They would use the ranked list of pages to decide which pages should be reviewed and updated first.

### Action

For the highest-priority pages, the SEO team may:

- Update outdated content
- Improve keyword optimization
- Refresh statistics and references
- Improve internal linking
- Rewrite titles or meta descriptions

### Cost of a Wrong Recommendation

If a page is incorrectly identified as needing a refresh, valuable editor time is wasted on pages that may already be performing well.

If an important declining page is missed, it may continue losing impressions, clicks, and organic traffic before it receives attention.

Because both types of errors have different consequences, the model should support prioritization rather than make automatic decisions.
